```python
# Exploring the distribution of age, avg_glucose_level, and bmi.

plt.figure(figsize=(12, 8))
plt.subplot(3,3,1)
plt.hist(df2['age'],edgecolor='black')
plt.title('Age Distribution')
plt.ylabel('Frequecy')
plt.xlabel('Age')

plt.subplot(3,3,2)
plt.hist(df2['avg_glucose_level'],color='skyblue',edgecolor='black')
plt.title('avg_glucose_level distribution')
plt.xlabel('Avg. glucose level')


plt.subplot(3,3,3)
plt.hist(df2['bmi'],color='lightblue',edgecolor='black')
plt.title('bmi Distribution')
plt.xlabel('Body Mass Index');
```

```python
df2[["age","hypertension", "heart_disease", "bmi","stroke","avg_glucose_level" ]].corr().round(3)
```

```python
# Visualizing the distribution of strokes in the dataset.
plt.figure(figsize=(3, 3))
df2.groupby('stroke')['stroke'].count().plot.bar(edgecolor='black');
plt.legend()
plt.title('Stroke distribution')
plt.ylabel('No. of people with stroke');
```

```python
# Analyzing the distribution of gender, work_type, and smoking_status with respect to strokes

plt.figure(figsize=(12, 10))

plt.subplot(3,3,1)
df2.groupby('gender')['stroke'].sum().plot.bar(edgecolor='black');
plt.legend()
plt.title('Stroke dist. vs to Gender')
plt.ylabel('No. of people with stroke')

plt.subplot(3,3,2)
df2.groupby('work_type')['stroke'].sum().plot.bar(edgecolor='black',color='skyblue');
plt.legend()
plt.title('Stroke dist. vs to Work Type')

plt.subplot(3,3,3)
df2.groupby('smoking_status')['stroke'].sum().plot.bar(edgecolor='black',color='lightblue');
plt.legend()
plt.title('Stroke dist. Vs smoking_status');
```

```python
# Select the columns of interest
selected_columns = ["age", "hypertension", "heart_disease", "bmi", "stroke", "avg_glucose_level"]

# Create a correlation matrix
correlation_matrix = df2[selected_columns].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

plt.title("Correlation Heatmap")
plt.show()
```
